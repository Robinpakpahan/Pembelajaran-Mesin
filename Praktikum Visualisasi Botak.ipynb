## Diagram Tingkat Pendidikan ##

```python
# Mengatur gaya visualisasi
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.countplot(x=df["pendidikan"], palette="viridis", order=df["pendidikan"].value_counts().index)
plt.title("Distribusi Tingkat Pendidikan")
plt.xlabel("Tingkat Pendidikan")
plt.ylabel("Jumlah")
plt.show()
```

#### Jika ada tingkat pendidikan yang dominan, seperti SMA atau S1, maka mayoritas responden memiliki latar belakang tersebut. Kesenjangan besar antar kategori menunjukkan tingkat pendidikan yang lebih tinggi lebih umum. Jika distribusi merata, maka ada keberagaman dalam pendidikan. Selain itu, hubungan antara pendidikan dengan gaji, pekerjaan, atau kebotakan bisa dianalisis lebih lanjut untuk melihat pola yang signifikan. ####

## Diagram Perokok Vs Non Perokok ##

```python
plt.figure(figsize=(6, 6))
df["is_merokok"].value_counts().plot.pie(autopct='%1.1f%%', labels=["Tidak Merokok", "Merokok"], colors=["#66b3ff", "#ff9999"])
plt.title("Proporsi Perokok vs Non-Perokok")
plt.ylabel("")  # Menghilangkan label sumbu Y
plt.show()
```

#### Jumlah perokok lebih besar dibanding org yang tida, maka kebiasaan merokok umum di dataset, sedangkan dominasi non-perokok menunjukkan sebaliknya. Jika perbandingan hampir seimbang, maka distribusinya merata. Bisa dianalisis apakah merokok berpengaruh terhadap kebotakan dan apakah kebiasaan ini lebih umum pada jenis kelamin atau kelompok usia tertentu. ####

## Diagram Umur memengaruhi Kebotakan ##

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df["umur"], y=df["botak_prob"], marker="o", color="red")
plt.title("Hubungan Umur dengan Probabilitas Kebotakan")
plt.xlabel("Umur")
plt.ylabel("Probabilitas Kebotakan")
plt.show()
```

#### Kebotakan meningkat seiring bertambahnya usia, maka faktor usia berpengaruh. Namun, jika ada fluktuasi, maka faktor lain juga terlibat. Jika peningkatan signifikan terjadi setelah usia tertentu, seperti 40 tahun, maka usia kemungkinan menjadi faktor utama. Dan bisa dilihat apakah laki-laki lebih rentan terhadap kebotakan dibandingkan perempuan serta peran faktor lain seperti merokok atau penggunaan sampo. ####
